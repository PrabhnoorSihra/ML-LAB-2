```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import random
from scipy.stats import norm
```

```python
random.seed(1)
x = scipy.stats.norm.rvs(0, 1, 100)
```

```python
eps = scipy.stats.norm.rvs(0, 0.25, 100)
```

```python
y = -1 + 0.5 * x + eps
print(y)
```

```python
%matplotlib inline
plt.scatter(x,y)
plt.show()
```

```python
XYdata = {'X' : x , 'Y':y}
df = pd.DataFrame(XYdata)

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
import statsmodels.formula.api as sm
```

```python
results = sm.ols('Y ~ X', df).fit()
print(results.__dict__)
#help(results)
print ('.......... Results .......................')
print(results.summary())
```

```python
%matplotlib inline
y_true = df.Y.values.copy()
#Predicted values of y from the above linear regression model
y_predicted = results.predict(df.X)
#Population values of Y
y_population = -1 + 0.5*df.X.values
#plot the actual data points
plt.plot(df.X, y_true, 'o', label='Data')
#plot the calculated line using linear regression
plt.plot(df.X, y_predicted, 'r-', linewidth=3, label='Fit')
#plot the population line
plt.plot(df.X, y_population, 'g--', linewidth=1, label='Population')
#Display the axes labels and legend
plt.xlabel('X')
plt.ylabel('mpg')
plt.legend(loc='best')
print('Parameters: ', results.params)
print('Standard errors: ', results.bse)
print('R-square: ', results.rsquared)
```

```python
x_1 = scipy.stats.norm.rvs(0,1,100)
eps_1 = scipy.stats.norm.rvs(0, 0.05, 100) #std _dev=0.05
y_1 = -1 + 0.5*x_1 + eps_1
x_2, y_2 = zip(*sorted(zip(x_1, y_1)))
XYdataLessNoise = {'X' : x_2 , 'Y':y_2}
#create a dataframe
dfLessNoise = pd.DataFrame(XYdataLessNoise)
#Perform linear regression on the Less noisy data
resultsLessNoise = sm.ols('Y ~ X', dfLessNoise).fit()
print(resultsLessNoise.__dict__)
#Plot the data
plt.plot(dfLessNoise.X, dfLessNoise.Y, 'o', label='Data')
#plot the prediction from the linear model
plt.plot(dfLessNoise.X, dfLessNoise.Y, 'r-', label='Fit')
plt.xlabel('X'), plt.ylabel('Y'), plt.legend()
print('Parameters: ', resultsLessNoise.params)
print('Standard errors: ', resultsLessNoise.bse)
print('R-square: ', resultsLessNoise.rsquared)
```

```python
x_2 = scipy.stats.norm.rvs(0, 1, 100)
eps_2 = scipy.stats.norm.rvs(0, 1.25, 100)
y_2 = -1 + 0.5*x_2 + eps_2
XYdataMoreNoise = {'X':x_2, 'Y':y_2}
#create a dataframe
dfMoreNoise = pd.DataFrame(XYdataMoreNoise)
#Perform linear regression on the Less noisy data
resultsMoreNoise = sm.ols('Y ~ X', dfMoreNoise).fit()
print(resultsMoreNoise.summary())
#Plot the data
plt.plot(dfMoreNoise.X,dfMoreNoise.Y, 'o', label='Data')
#plot the prediction from the linear model
plt.plot(dfMoreNoise.X, dfMoreNoise.Y, 'r-', label='Fit')
plt.xlabel('X'), plt.ylabel('Y')
print('Parameters: ', resultsMoreNoise.params)
print('Standard errors: ', resultsMoreNoise.bse)
print('R-square: ', resultsMoreNoise.rsquared)
#Plot the LOWER and UPPER 95% confidence interval lines
from statsmodels.sandbox.regression.predstd import wls_prediction_std
_, upper,lower = wls_prediction_std(resultsMoreNoise)
plt.plot(dfMoreNoise.X, upper, '--k', label="Upper")
plt.plot(dfMoreNoise.X, lower, '--g', label="Lower")
plt.legend(loc='best')
```
